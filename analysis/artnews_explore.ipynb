```python
# External modules
import pandas as pd
import matplotlib.pyplot as plt
# Project modules
import artnews
```

```python
# Cortei o final de alguns termos para pegar derivações (ex. impressionism, impressionist...)
styles = ['Abstract', 'Fine Art', 'Modern', 'Abstract Expressionis',
          'Expressionis', 'Figurati', 'Impressionis', 'Realis',
          'Conceptual', 'Minimalis', 'Portrait', 'Pop Art', 'Surrealis',
          'Illustrati', 'Art Deco', 'Street Art', 'Photorealis', 'Folk',
          'Cubis', 'Documentar', 'Dada']
```

```python
articles = []
for topic in artnews.topics:
    articles.append(artnews.get_articles(topic))
```

```python
def style_time_series():
    time_series = pd.DataFrame(columns=styles)

    # Iterate over the news entries
    for entry in articles:
        title = entry['title'] + '' + entry['summary']
        try:
            timestamp = pd.to_datetime(entry['timestamp'])
        except:
            timestamp = pd.to_datetime('29 June 2023 20:01:54')
        
        # Extract the month and year from the timestamp
        month_year = timestamp.strftime('%Y-%m')
        
        # Iterate over the styles and check if they appear in the title
        for style in styles:
            if style.lower() in title.lower():
                # Increment the count for the corresponding style and month_year
                time_series.loc[month_year, style] = time_series.get((month_year, style), 0) + 1

    # Fill missing months with zeros and sort the DataFrame by the index (month_year)
    time_series = time_series.fillna(0).sort_index()

    return time_series
```
